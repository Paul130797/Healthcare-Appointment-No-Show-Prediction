```python
import pandas as pd
```

```python
#load the dataset
df=pd.read_csv('D:\\Internship Elevate labs\\Project 1\\noshowappointments.csv')
```

```python
#Basic cleanup
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df=df.rename(columns={'hipertension':'hypertension','handcap':'handicap','no_show':'no_show'})
```

```python
df.head()
```

```python
#convert dates

df['scheduledday'] = pd.to_datetime(df['scheduledday'])
df['appointmentday'] = pd.to_datetime(df['appointmentday'])
```

```python
#feature engineering

df['appointment_weekday'] = df['appointmentday'].dt.day_name()
df['scheduled_weekday'] = df['scheduledday'].dt.day_name()
df['no-show']=df['no-show'].map({'Yes':1,'No':0})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
#Drop irrelevant columns
columns_to_drop = ['PatientId', 'AppointmentId', 'ScheduledDay', 'AppointmentDay', 'Neighbourhood']
```

```python
#define features and target 

features=['age','sms_received','hypertension','diabetes','alcoholism','handicap','no-show']
x=df[features]
y=df['no-show']
```

```python
#train-test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#model 
model = DecisionTreeClassifier(max_depth=5, random_state=42)
model.fit(X_train, y_train)
```

```python
#Evaluate the model
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
